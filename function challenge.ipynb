```python
age = int(input("What is your age?\n"))
def year_born(num):
    return 2022 - num

print(year_born(age))
```

```python
import random

number = int(input("input a random number"))
def ran_num(max):
    return random.randint(1, max)

print(ran_num(number))
```

```python
month = input("enter a month")
def season(mon):
    mon = mon.upper()
    if mon == "MARCH" or mon == "APRIL" or mon == "MAY":
        return "Spring"
    elif mon == "JUNE" or mon == "JULY" or mon == "AUGUST":
        return "Summer"
    elif mon == "SEPTEMBER" or mon == "OCTOBER" or mon == "NOVEMBER":
        return "Fall"
    else:
        return "Winter"

print(season(month))
```
